```python
import ipyplot
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Images.csv",sep='\t')
```

```python
pre_prefix = 'Images/PreRequest/'
images = [pre_prefix + i for i in data['pre_request_file_name'].values]
post_prefix = 'Images/PostRequest/'
post_images = [post_prefix + i for i in data['post_request_file_name'].values]
images = np.concatenate((images,post_images))

labels = data['id'].values
labels = np.concatenate((labels,labels))
request = data['request_text'].values
request = np.concatenate((request,np.empty(30,dtype = '<U10')))

ipyplot.plot_class_tabs(images, labels, img_width=300,custom_texts=request)
```
