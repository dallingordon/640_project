```python
!accelerate launch --mixed_precision="fp16" train_text_to_image.py \
  --pretrained_model_name_or_path="CompVis/stable-diffusion-v1-4" \
  --train_data_dir="../../stable_edit/reddit_api/" \
  --use_ema \
  --resolution=512 --center_crop --random_flip \
  --train_batch_size=1 \
  --gradient_accumulation_steps=4 \
  --gradient_checkpointing \
  --max_train_steps=100 \
  --learning_rate=1e-05 \
  --max_grad_norm=1 \
  --lr_scheduler="constant" --lr_warmup_steps=0 \
  --output_dir="reddit_api_out"
```

```python
from diffusers import StableDiffusionPipeline

model_path = "reddit_api_out"
pipe = StableDiffusionPipeline.from_pretrained(model_path, torch_dtype=torch.float16)
pipe.to("cuda")

image = pipe(prompt="yoda").images[0]
image.save("yoda-pokemon.png")
```

#Train Dreambooth with Reddit finetuned Stable diffusion model

```python
!python ../dreambooth/train_dreambooth.py \
  --pretrained_model_name_or_path="reddit_api_out/"  \
  --instance_data_dir="dogleash/" \
  --output_dir="dreembooth_dogleash/" \
  --instance_prompt="<dogleash>" \
  --resolution=512 \
  --train_batch_size=1 \
  --gradient_accumulation_steps=1 \
  --learning_rate=5e-6 \
  --lr_scheduler="constant" \
  --lr_warmup_steps=0 \
  --max_train_steps=300
```

```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "dreembooth_dogleash/"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16).to("cuda")

```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
prompt = "<dogleash> in a bollywood movie"

for i in range(0,10):
  image = pipe(prompt, num_inference_steps=50, guidance_scale=7.5).images[0]
  image = np.array(image)
  plt.imshow(image)
  plt.show()
```
